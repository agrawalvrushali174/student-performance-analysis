```python
import pandas as pd
import numpy as np

df = pd.read_csv('Student_performance_data__.csv')
print(df.shape)
df.head() 
```

```python
df['GradeClass'] = df['GradeClass'].astype(int)

categorical_cols = ['Gender', 'Ethnicity', 'ParentalEducation', 'ParentalSupport',
                    'Tutoring', 'Extracurricular', 'Sports', 'Music', 'Volunteering']
for col in categorical_cols:
    df[col] = df[col].astype('category')

print(df.dtypes) 
```

```python
df['Gender'] = df['Gender'].map({0: 'Male', 1: 'Female'})
df['Tutoring'] = df['Tutoring'].map({0: 'No', 1: 'Yes'})
df['Extracurricular'] = df['Extracurricular'].map({0: 'No', 1: 'Yes'})
df['Sports'] = df['Sports'].map({0: 'No', 1: 'Yes'})
df['Music'] = df['Music'].map({0: 'No', 1: 'Yes'})
df['Volunteering'] = df['Volunteering'].map({0: 'No', 1: 'Yes'})
df.head()
```

```python
def expected_grade(gpa):
    if gpa >= 3.5: return 0
    elif gpa >= 3.0: return 1
    elif gpa >= 2.5: return 2
    elif gpa >= 2.0: return 3
    else: return 4

df['expected_grade'] = df['GPA'].apply(expected_grade)
mismatches = df[df['expected_grade'] != df['GradeClass']]
print(f"Inconsistent rows: {len(mismatches)}")
```

```python
df['StudyTimeWeekly'] = df['StudyTimeWeekly'].round(2)
df['GPA'] = df['GPA'].round(4)
df[['StudyTimeWeekly', 'GPA']].head()
```

```python
print("Duplicate IDs:", df['StudentID'].duplicated().sum())
expected_ids = set(range(1001, 3393))
actual_ids = set(df['StudentID'])
missing_ids = expected_ids - actual_ids
print(f"Missing IDs from sequence: {len(missing_ids)}")
```

```python
print("Age valid:", df['Age'].between(15, 18).all())
print("GPA valid:", df['GPA'].between(0.0, 4.0).all())
print("Absences valid:", df['Absences'].between(0, 30).all())
print("StudyTime valid:", df['StudyTimeWeekly'].between(0, 20).all())
```

```python
zero_gpa = df[df['GPA'] == 0.0]
print(f"Students with GPA = 0: {len(zero_gpa)}")
zero_gpa[['StudentID', 'Absences', 'StudyTimeWeekly', 'GPA', 'GradeClass']]
```

```python
df.drop(columns=['expected_grade'], inplace=True)  # remove helper column
df.to_csv('Student_performance_cleaned.csv', index=False)
print("✅ Cleaned file saved!")
```
